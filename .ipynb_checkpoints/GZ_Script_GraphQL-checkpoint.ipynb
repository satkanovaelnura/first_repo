```python
import requests
import json
import pandas as pd
import time 

hernya = "000140001615"

# Создайте список параметров, по которым вы хотите получить данные
parameters = [
    "contractNumberSys",
    "Supplier{nameRu}",
    "descriptionRu",
    "FaktTradeMethods{nameRu}",
    "faktSum",
    "contractSumWnds",
    "RefSubjectType{nameRu}"
]

df_12 = pd.DataFrame()

def send_graphql_request(customer_bin, parameter, after=None):
    global df_12  # Объявляем df_12 как глобальную переменную

    url = "https://ows.goszakup.gov.kz/v3/graphql/query"
    headers = {
        "Authorization": "Bearer 36e52c17fd04fd3ebd3222319c4f289b",
        "Content-Type": "application/json"
    }

    filter_params = f'{{customerBin: "{customer_bin}", finYear: 2023}}'

    graphql_query = {
        "query": f'query GetContracts {{Contract(limit: 200, filter: {filter_params}) {{ {parameter} }}}}'
    }

    try:
        response = requests.post(url, headers=headers, json=graphql_query, verify=False)
        response.raise_for_status()
        data = response.json()

        contracts = data.get("data", {}).get("Contract", [])

        if contracts is not None:
            new_df = pd.DataFrame(contracts)
            new_df.rename(columns={parameter: parameter}, inplace=True)
            df_12 = pd.concat([df_12, new_df], axis=1)

    except requests.exceptions.RequestException as e:
        print(f"Ошибка при выполнении запроса для {customer_bin}: {e}")

# Список данных
customer_bin_list = [hernya]

# Повторяем запросы для каждого параметра и каждого значения в списке
for parameter in parameters:
    for customer_bin in customer_bin_list:
        send_graphql_request(customer_bin, parameter)
        time.sleep(1)  # Добавьте задержку в несколько секунд перед следующим запросом

print(df_12)
print("END")
```
